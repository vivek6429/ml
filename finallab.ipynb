```python
#Which location having maximum death cases for male and female and Print all types of symptom.
import pandas as pd
entiredf=pd.read_csv("COVID1.csv")
```

```python
print(df.columns)
# last  columns are empty so dropping them
df=df.drop(['Unnamed: 3','Unnamed: 21', 'Unnamed: 22',
       'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26'],axis=1)
```

```python
#Which  location  having maximum  death cases  for  --male and female ---  and Print all types of symptom.

# df.country.unique
print(df['death'].unique())
print()
print(entiredf['death'].unique())
# we have some entries with wrong formattings
print(df['death'].describe())
# gona drop those rows 
```

```python
#['2/14/2020', '2/26/2020', '2/13/2020', '2/28/2020', '2/27/2020', '2/25/2020', '2/23/2020','2/24/2020', '2/22/2020', '02/01/20' ,'2/19/2020' , '2/21/2020']
#     df.drop(indexname,inplace=True)
df=df.loc[df.death.isin(['0','1'])]
df.gender=df.gender.astype(str)
df.gender.unique()
df=df.loc[df.gender.isin(['male','female'])]
df.gender.unique()
df.death=df.death.astype(int)
```

```python
df_small=df[["location","gender","death"]]
print(df_small.death.sum())
# we now have the data grouped by province
df_small=df_small.groupby(["location","gender"])
df_small.head()
#df_small.sum().values
# pd.DataFrame(df_small).to_excel('smalldb.xlsx', sheet_name='new_sheet_name')
```

```python
s=df_small
# find location with max  deaths
s.head()
# import pandas as pd
# import openpyxl
# s.to_excel('pandas_to_excel_no_index_header.xlsx', index=False, header=False)
```

```python
# s.loc[s['death'].idxmax()]
print("Female max death location:")
print(s.xs('male',level='gender').idxmax().values)
print()
print("Male max death location:")
print(s.xs('female',level='gender').idxmax().values)
```
